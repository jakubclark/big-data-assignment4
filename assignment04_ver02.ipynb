# Question 6: How does the neighborhood of the home affect its sale price

```python


for zip_code in set(list(df['ZIPCODE'])):
    if zip_code:
        print("-"*20)
        print("\nData frame for zip code: ", zip_code)
        df_zipcode = df[df['ZIPCODE']==zip_code]
        print(df_zipcode.groupby(['ROOMS']).agg({'PRICE':['min','max']}))
```

# Question 3: How does the age of the home affect its selling price?

```python
room_num_lst = set(list(df['ROOMS']))
```

```python

for room_num in room_num_lst:
    print('-'*50)
    
    fig = plt.figure()
    
    if room_num < 19:
        print("Age of homes with {} room(s) and their selling price".format(room_num))
        fig.suptitle('the age of the home with {} rooms affect its selling price'.format(room_num), fontsize=20)
        df_age = df[df['ROOMS']==room_num][['AYB', 'PRICE']].groupby('AYB').agg('mean').sort_values('AYB')
        df_age.index = [int(x) for x in df_age.index]
    else:
        print("Age of homes with {} room(s) or more and their selling price".format(room_num))
        fig.suptitle('the age of the home with {} rooms or more affect its selling price'.format(room_num), fontsize=20)
        df_age = df[df['ROOMS']>=room_num][['AYB', 'PRICE']].groupby('AYB').agg('mean').sort_values('AYB')
        df_age.index = [int(x) for x in df_age.index]

    
    
    plt.plot(df_age.index, df_age['PRICE'], 'b') 
    plt.xlabel('AYB') 
    plt.ylabel('PRICE') 
    plt.xticks(np.arange(min(df_age.index), max(df_age.index)+1, 25))#((max(df_age.index)-min(df_age.index))+1)//10)) 
    plt.legend()
    plt.show()
    if room_num >= 19:
        break
```

# Question 4: How did the sale price of homes change over time?

```python
nbhd_num_lst = set(list(df['ASSESSMENT_NBHD'])) 
```

```python
plt.rcParams['figure.figsize'] = [15, 5]

df_age = df[['AYB', 'PRICE']].groupby('AYB').agg('mean').sort_values('AYB')

df_age.index = [int(x) for x in df_age.index]

fig = plt.figure()
fig.suptitle('The sale price change over time', fontsize=20)
plt.plot(df_age.index, df_age['PRICE'], 'b') 
plt.xlabel('Year house is built') 
plt.ylabel('PRICE (in millions)') 
plt.xticks(np.arange(min(df_age.index), max(df_age.index)+1, 25)) 
plt.legend()
plt.show() 
#fig = hst3.get_figure()
fig.savefig('price_freq_.png', dpi=300)
fig.savefig('price_freq_.pdf')
```
